```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('train.csv')
df.sample(10)
```

# Drawing plots using matplotlib

```python

plt.hist(df['Survived'] )
plt.xticks([0,1])
plt.xlabel('survival')
plt.ylabel('count')
plt.title('Counting The Survived People')
plt.show()
```

# Drwaing same plot using seaborn

```python
sns.countplot(x=df['Survived'])

```

# Doing the same thing through  pandas .values_count()

```python
x = df['Survived'].value_counts()
x.plot(kind='bar')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Survival Count')
plt.xticks([0, 1], ['Not Survived', 'Survived'], rotation=0)
plt.show()

```

```python
df['Survived'].value_counts()
plt.bar(x.index, x.values)
plt.xticks([0,1] ,['Died','Survived'])
plt.xlabel('survival')
plt.ylabel('count')
```
